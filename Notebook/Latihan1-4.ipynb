# Latihan 1
Buatlah program untuk menghitung banyaknya transaksi yang terjadi di supermarket X yang berada di negara Kanada selama bulan November 2016. Gunakan data dari file Excel transaksi belanja (Sample - Superstore.xls)

## Mempersiapkan Data

```
import pandas as pd
import numpy as np
import os
```

```
# Loading data excel ke dalam notebook dengan pandas
df = pd.read_excel('../Data/Sample%20-%20Superstore.xls')
```

```
df.head()
```

## Seleksi Data

```
# Menampilkan 10 random sample dari dataframe
df['Order ID'].sample(n=10)
```

```
order = df['Order ID'].apply(lambda x : x[:2])
```

```
order.head()
```

```
order.value_counts()
```

## Seleksi Order di Canada

```
# Seleksi order CA untuk Canada
df = df[order == 'CA']
```

```
# Reset Index dataframe
df.reset_index(drop=True,inplace=True)
```

```
df.tail()
```

## Seleksi Order November 2016

```
import datetime
```

```
# Filter tanggal 1-11-2016 sd 30-11-2016 aka November 2016
df = df[df['Order Date'] >= pd.Timestamp(2016,11,1)]
df = df[df['Order Date'] <= pd.Timestamp(2016,11,30)]
```

```
# Reset dataframe index
df.reset_index(drop=True,inplace=True)
```

```
# Hasil akhir dataframe, order di Canada pada bulan November 2016
# Terdapat 313 order dari Canada pada bulan November tahun 2016
df
```

```
# Validasi data
# Bernilai True jika dibawah 360 (30 x 12) Jumlah total hari dalam satu tahun
len(df3['Order Date'].unique()) < 360
```
