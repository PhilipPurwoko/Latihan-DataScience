# Latihan 5
Membuat rekap total keuntungan di supermarket Canada dan US pada tahun 2016 dan 2017

```
df5 = pd.read_excel('../Data/Sample%20-%20Superstore.xls')
```

```
df5.head()
```

```
# Select 2016
df_2016 = df5[df5['Order Date'] >= pd.Timestamp(2016,1,1)][df5['Order Date'] <= pd.Timestamp(2016,12,31)]

# Select 2017
df_2017 = df5[df5['Order Date'] >= pd.Timestamp(2017,1,1)][df5['Order Date'] <= pd.Timestamp(2017,12,31)]
```

```
print('Total profit dari seluruh supermarket di Canada dan US pada tahun 2016 dan 2017')
print(f'2016 = ${np.float64(sum(df_2016["Profit"]))}')
print(f'2017 = ${np.float64(sum(df_2017["Profit"]))}')
```

# Latihan 6
Kategori produk yang paling banyak diberikan diskon pada tahun 2014-2017 (semua data)

```
df6 = pd.read_excel('../Data/Sample%20-%20Superstore.xls')
```

```
df6.head(1).T
```

```
df6['Discount'].head()
```

```
df6['Category'].head()
```

```
category_count = {i:0 for i in df6['Category'].unique()}

for cat,dis in zip(df6['Category'],df6['Discount']):
    if dis != np.float64(0):
        category_count[cat] += 1

print(category_count)

# Kesimpulan :
# Kategori Office Supplies adalah kategori produk yang paling sering memberikan diskon selama tahun 2014-2017
```
